Introduction:

In the previous report we explored building efficiency and its relationship to specific building characterics. The previous report concluded that relative compactness has a moderate to strong correlation to building energy efficiency, as measured by heating and cooling load. Generally, as relative compactness increases so does building effeciency. Relative compactness is the ratio of a buildings internal volume to external surface area (where heat transfers). The more compact the building (i.e. the less surface area exposed to internal volume), the more energy efficient the building. Surface area values are in the dataset, but it also includes roof area which is also in the dataset examined. Volume, not directly in the dataset, is related to height which is in the dataset.

In this report we will build on the previous report and examine the key components of relative compactness. We will explore the underlying characteristics of surface area, volume, and their relationship to building energy efficiency. 

Summary:

Increasing relative compactness does increases efficiency, and diving deeper into the relationship shows that variables that make up relative compactness as well as other building parameters influence efficiency. From the analysis overall height, roof area, wall area, and glazing area have a positive relationship to efficiency. The Discussion section and Figures 1-7 build to and support this assertion.

Discussion:

1) The dataset provided by UCI has two distinct values of height - 3.5 and 7.0. Overall height has a positive relationship to heating and cooling load. In this dataset, heights of 7.0 have a greater efficiencies than heights of 3.5. This is evident by the Figure 1. The dataset only included two values of height, so we can only conclude that 7.0 is more efficient than 3.5, but it still fits the original conclusion that as relative compactness increases so does efficiency. As mentioned, height is a component of volume and volume is component of relative compactness. As height increases so does volume (holding all other dimesions constant). If the increase of height results in greater compactness (more volume than surface area added), efficiency increases. 

2) Additionally, from examining roof area we see that roof areas of 220.5 are less efficient than roof areas below that mark. This is evident by Figure 2. While there are only  four factors of roof area in the dataset,  they are ordinal and there is a large gap in continuity in the values provided from 147 to 220.5. This drastic increase in roof area and associated loss in efficiency also fits the original conclusion. Roof area is a component of external surface area from which energy dissipates/transfers to the enviornment. As roof area increases so does external surface area. If roof area increases such that surface area added increases more than internal volume the relative compactness decreases and efficiency decreases.

3) Examining surface area and its relationship to one of the efficiency measures (heating load) shows that in general as external surface area increases efficiency decreases. Figure 3 shows this relationship. If the increase in surface area results in a decrease in compactness - greater proportion of surface area added than volume - and decrease in heating load. Figure 3 also shows that roof areas and heights that are 220.5 and 3.5, respectively, are less efficient and for other values (not 220.5 and 3.5) for these variables show an another possible relationship to heating load. 

4) Figure 4 shows the inverse relationship between relative compactness and heating load (efficiency). As relative compactness increases so does energy efficiency. Again, heights and roof areas mentioned in #3 show a significant impact on the positive relationship. For heights and roof areas not 220.5 and 3.5, there seems to be another influencing variable driving efficiency. Additionally for values of relative compactness between 0.75 and 0.80 seem to have a slighly higher efficiencies than higher relative compactness values.

5) The slighly higher efficiences for the relative compactness values between 0.75 and 0.80 may be explained by wall area. Figure 5 shows that for heights of 7.0 and relative compactness between 0.75 and 0.80 seem to have higher average efficiencies than other compactness values. Wall area seems to have some impact on efficiency, which is confirmed by the correlation matrix presented in the previous report.

6) Adding glazing area to Figure 5 (shown in Figure 6), shows that glazing area has some impact on efficiency within dimension characteristic groupings. The more area that is glazed, the greater the efficiency compared to similar buildings without glazing. Again this is confirmed by the correlation matrix presented in the previous report.

7) Figure 7 illustrates #6 more clearly, that glazing area has a positive relationship with efficiency, especially within building dimension groupings (Wall Area and Height). The relationship is particulally clear for heights of 3.5.





Figure 1:

The following figure shows the relationship of cooling load to heating load. Not only is heating and cooling load strongly correlated, but the two factors of height show that, for the most part, overall heights of 3.5 have lower efficiencies than overall heights of 7.0. Note the grouping of red points at the lower end of the efficiency range.

```r
# get data
energydata <- read.csv(file = "EnergyEfficiencyData.csv", header = TRUE)
# set factors
facCol <- c("Overall.Height", "Orientation", "Glazing.Area.Distribution")
energydata[, facCol] <- lapply(energydata[,facCol], as.factor)

require(ggplot2)

ggplot(energydata, aes(Cooling.Load, Heating.Load)) + geom_point(aes(color = factor(Overall.Height)), alpha = 0.3) + 
  xlab('Cooling Load') + ylab('Heating Load') + 
  ggtitle('Cooling vs. Heating Load with Height')
```

Figure 2:

The following figure shows that the majority of buildings with roof areas of 220.5 are less efficient than buildings with less roof areas. For roof areas of 147 and below, there is no discernable (visible) difference related to efficiency. Note the purple points (220.5) at the lower end of the efficiency range.

```r
ggplot(energydata, aes(Cooling.Load, Heating.Load)) + geom_point(aes(color = factor(Roof.Area)), alpha = 0.3) + 
  xlab('Cooling Load') + ylab('Heating Load') + 
  ggtitle('Cooling vs. Heating Load with Roof Area')
```

Figure 3:

The following figure shows that as surface area increases heating load decreases. Roof area and height seem to have a significant impact on the negative relationship. Note the purple circles.

```r
ggplot(energydata, aes(Surface.Area, Heating.Load)) + geom_point(aes(color = factor(Roof.Area),
                                                                             shape = factor(Overall.Height)), alpha = 0.3) + 
  xlab('Surface Area') + ylab('Heating Load') + 
  ggtitle('Surface Area vs. Heating Load by Roof Area and Height')
```

Figure 4:

The following figure shows that as relative compactness increases heating load increases. Roof area and height seem to have a significant impact on the positive relationship. Note the purple circles.

Also, there seems to be more efficiency (load) for compactness between .75 and .80 than greater values of compactness, which points to another influencing variable.


```r
ggplot(energydata, aes(Relative.Compactness, Heating.Load)) + geom_point(aes(color = factor(Roof.Area), size = Surface.Area,
                                                                             shape = factor(Overall.Height)), alpha = 0.3) + 
  xlab('Relative Compactness') + ylab('Heating Load') + 
  ggtitle('Relative Compactness vs. Heating Load by Roof area and Height')
```

Figure 5:

The following figure shows how wall area affects heating load with relative compactness. For heights of 7.0 and wall areas above 343.0, there is a slighly higher efficiency. 

```r
ggplot(energydata, aes(Relative.Compactness, Heating.Load)) + geom_point(aes(color = factor(Roof.Area), size = Surface.Area,
                                                                             shape = factor(Overall.Height)), alpha = 0.3) + 
  facet_grid(.~Wall.Area) +
  xlab('Relative Compactness') + ylab('Heating Load') + 
  ggtitle('Relative Compactness vs. Heating Load by Wall Area')
```

Figure 6:

The following figure shows how glazing area affects heating load. Note the size of the points within groupings. Smaller points, presumably less glazed area, are less efficient than more glazed area within groups.

```r
ggplot(energydata, aes(Relative.Compactness, Heating.Load)) + geom_point(aes(color = factor(Roof.Area), size = Glazing.Area,
                                                                             shape = factor(Overall.Height)), alpha = 0.3) + 
  facet_grid(.~Wall.Area) +
  xlab('Relative Compactness') + ylab('Heating Load') + 
  ggtitle('Relative Compactness vs. Heating Load by Wall Area')
```

Figure 7:

The following figure shows that glazing area increases efficiency compared to simlar buildings without or with less glazing. Note the purple dots are higher and further right than other colors within groups.

```r
ggplot(energydata, aes(Cooling.Load, Heating.Load)) + geom_point(aes(color = factor(Glazing.Area)), alpha = 0.3) + 
  xlab('Cooling Load') + ylab('Heating Load') + 
  facet_grid(Overall.Height~Wall.Area) +
  ggtitle('Cooling vs. Heating Load with Glazing Area by Wall Area and Overall Height')
```
