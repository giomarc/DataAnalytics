## Introduction

In this data exploration we look at several variables that affect the energy efficiency of buildings, as measured by the cooling load and heating load. An assumption of this analysis is that higher heating and cooling loads means higher efficiencies, this may not be the case.

## Summary

The heating and cooling loads of buildings depends on the relative compactness, surface area, roof area, and overall height of the building. Relative compactness is an all encompassing measure of building compactness building or the ratio of internal volume to external surface area. The measures of surface area, roof area, and overall height are likely used in the calculation of relative compactness since there is high correlation, either negative or positive, between them, as exhibited by Figure 2. More compact buildings seem to positively influence loads. Figure 3 shows that overall height impacts load - the higher of the two factors given the dataset exhibit higher loads. Heights of 7.5 have a much higher efficiency or load than those with 3.5. Figure 4 shows roof area also impacts load - the greater the roof area the lower the load or efficiency. 

The main conclusion is that the more compact a building the more efficient. Heights of 7.5 also show greater efficiency than heights of 3.5. Additionally, minimal exterior surface area and roof area leads to higher efficiency because less energy is lost through the extremities of the building.  



### Get data and a high level understanding

1) get the data from a csv
2) based on the description of the data from the source repo, Orientation and Glazing Area Distribution are categorical variables but are integers in the data. Change them to factors
3) get the structure

```r
# read csv
energydata <- read.csv(file = "EnergyEfficiencyData.csv", header = TRUE)
#head(energydata)
# change ints to factors
facCol <- c("Orientation", "Glazing.Area.Distribution")
energydata[, facCol] <- lapply(energydata[,facCol], as.factor)
# understand structure and summary stats
# str(energydata)
# summary(energydata)
```

### Figure 1 - Scatterplot Matrix

To get a high level understanding of the data we'll start with a scatter plot. A few things stand out:
1) Relative compactness is inversly related to surface area (top left (4x4) boxes)
2) As Relative compactness increases so does heating and cooling load (row 9&10 column 1)
3) As Surface area increases, heating and cooling load decreases (row 9&10, column 1)
4) There seems to be a single roof area that has particularly low heating and cooling load (row 9&10, column 4)
5) Of the two values in overall heights the greater value has higher loads (row 9&10, column 5)

We'll examine these relationships further.


```r
# correlation scatterplot.
require(car)
scatterplotMatrix(x = energydata, var.labels = colnames(energydata), diagonal = c("histogram"))
```

### Figure 2 - Correlation Matrix

The correlation matrix reveals some interersting takeaways from high correlations:
1) Roof area is strongly correlated with surface area, presumably because it is included in surface area.
2) Relative compactness is inversly correlated to surface area and roof area
3) Overall height is inversely correlated to surface area and roof area, but positively correlated to relative compactness
4) Heating and cooling load is negatively correlated to roof area and surface area, but positively related to relative compactness and overall height. 
5) Heating and cooling load are correlated to each other.

Since roof area and overall heights seem to have factors that affect load we'll next look at boxplots for these values against heating load.


```r
require(corrplot)
energydataSub <- energydata[, -which(names(energydata) %in% facCol)]
col <- colorRampPalette(c("#BB4444", "#EE9988", "#FFFFFF", "#77AADD", "#4477AA"))
R <- cor(x = energydataSub, method = c("pearson"))
#print(R)

# correlation plot code based on article: 
# http://www.sthda.com/english/wiki/visualize-correlation-matrix-using-correlogram
source('~/Dropbox/UW_DS/MethodsDS/DataAnalytics/cormtest.R')
p.mat <- cor.mtest(energydataSub)
#head(p.mat)

corrplot(R, method="color", col=col(200),  
         type="lower", order="hclust", 
         addCoef.col = "black", # Add coefficient of correlation
         tl.col="black", tl.srt=45, #Text label color and rotation
         # Combine with significance
         p.mat = p.mat, sig.level = 0.05, insig = "blank", 
         # hide correlation coefficient on the principal diagonal
         diag=FALSE 
)
```

### Figure 3 - Overall Height vs. Heating Load

The following boxplot reveals that heights of 7.5 have significantly (not statisticall, which is unknown from this analysis) higher heating loads than heights of 3.5. This fits correlation values we saw in Figure 2.

```r
require(ggplot2)
ggplot(energydata, aes(x = factor(Overall.Height), y = Heating.Load)) + geom_boxplot()
```

### Figure 4 - Roof Area vs. Heating Load

The following box plot reveals that for roof areas that are 220.5, heating load is much less than the other lower factors of roof area. This the high level conclusion that surface area, where roof area is included diminishes load/efficiency. 

```r
ggplot(energydata, aes(x = factor(Roof.Area), y = Heating.Load)) + geom_boxplot()
```
