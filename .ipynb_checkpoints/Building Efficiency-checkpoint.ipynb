Building Efficency: Exploration - Pranav Shah

In this data exploration we look at several variables that affect the energy efficiency of buildings, as measured by the cooling load and heating load. 

Bottom line

The heating and cooling loads of buildings depends on the relative compactness of the building. The more compact a building the more efficient. Additionally, minimal exterior surface area leads to higher efficiency because less energy is lost through the extremities of the building. The figures below walk through how we reached this conclusion. 



Get data and a high level understanding

1) get the data from a csv
2) based on the description of the data from the source repo, Orientation and Glazing Area Distribution are categorical variables but are integers in the data. Change them to factors
3) get the structure
4) summary statistics

```r
# read csv
energydata <- read.csv(file = "EnergyEfficiencyData.csv", header = TRUE)
head(energydata)
# change ints to factors
facCol <- c("Orientation", "Glazing.Area.Distribution")
energydata[, facCol] <- lapply(energydata[,facCol], as.factor)
# understand structure and summary stats
str(energydata)
summary(energydata)
```

From the above preliminary analysis, something was a bit strange about a few of the numeric dimesion variables (i.e. relative compactness, surface area, etc.). There seemed to be repetive numbers in many of the observations. This makes sense since the UCI description claims these numbers were generated through a simulation. Next, we want to see what is unique through distributions/histograms.

Understand the distribution of data

1) understand what is unique

```r
# histogram all values are they really numerics? 
# find number of unique values in each column
lapply(lapply(energydata, unique), length)
```

Here we see that many of the dimesion variables do in fact have repetive numbers. For instance, wall area has 7 unique values for 768 observations. While this is interesting, the values are ordinal, meaning there is increasing quantity to the values at higher values. Orientation does not exhibit this even though the factor levels increase from 2 to 5. 

2) The histogram code below shows the distribution values. (warning: this will generate six histograms)

```r
# visual illustration
require(ggplot2)
histFun <- function(x)
{
  cnames <- names(x)
  for(i in seq_along(cnames)){
    print(ggplot(x, aes_string(x = cnames[i])) + geom_histogram())
  }
}

#remove factors
energydataSub <- energydata[, -which(names(energydata) %in% facCol)]
# create histograms 
histFun(energydataSub)
```

From the above histograms we can see that a few of the dimensional values are unifomally distributed, which may speak to the nature of how the simulated buildings were generated from code. 


Understand relationship of data to each other through a scatterPlot Matrix: first ran with all energydata then with factor variables removed because no obvious relationship existed between factors and other variables.


```r
# correlation scatterplot.
require(car)
scatterplotMatrix(x = energydataSub, var.labels = colnames(energydataSub), diagonal = c("histogram"))
```

In the above scatterplot matrix we can see several takeaways:
- As relative.compactness increases so does heating and cooling load
- Surface area is inversely related to relative compactness
- Heating and cooling load are highly related to each other


To see these represented in numbers we will make a correlation matrix with non-factor variables.

```r
require(corrplot)
col <- colorRampPalette(c("#BB4444", "#EE9988", "#FFFFFF", "#77AADD", "#4477AA"))
R <- cor(x = energydataSub, method = c("pearson"))
#print(R)

# correlation plot code based on article: 
# http://www.sthda.com/english/wiki/visualize-correlation-matrix-using-correlogram
source('~/Dropbox/UW_DS/MethodsDS/DataAnalytics/cormtest.R')
p.mat <- cor.mtest(energydataSub)
#head(p.mat)

corrplot(R, method="color", col=col(200),  
         type="lower", order="hclust", 
         addCoef.col = "black", # Add coefficient of correlation
         tl.col="black", tl.srt=45, #Text label color and rotation
         # Combine with significance
         p.mat = p.mat, sig.level = 0.05, insig = "blank", 
         # hide correlation coefficient on the principal diagonal
         diag=FALSE 
)
```

From the correlation matrix above we see that relative compactness and height are moderately to highly correlated to load. Relative compactness is the ratio between exterior surface area and internal volume. The more surface area the less compact a building, thus the high negative correlation between the two. Note that height has a relationship to volume, which is why relative compactness and height are strongly correlated. Additionally, surface area and roof area are inversely correlated to load. The more exterior surface area of a building the less efficient it is because more heat transfer occurs at the extremities of the building. Roof area is included in surface area therefore correlated. 

Conclusion:
Buildings that are most efficient seem to be the ones that are the most compact, meaning less exposed surface area.
