## Introduction

In this exercise we'll examine what is the best strategy for the gameshow *Let's Make a Deal*. In this game, there are three doors. One door has a hidden car that a contestant can win. The other two doors have a goat. The objective is ot win the car. To begin the producers place the car behind one of the three doors randomly and goats in the other two, which the contestant does not witness. Next the contestant picks one of three doors she thinks the car is behind. Next the host will remove a door with a goat, but not one a contestant has chosen, if done so originally. The two doors that remain contain a goat and a car. The host then gives the contestant the option to switch or stay with their original choice. Before starting the game, should the contestant have an always switch mentality or always stay? Which choice increases your probability of winning the car? We use a simulation to determine which strategy is best. 

## Summary

After running 100,000 trials, the best strategy is to have an always switch mentality. Based on the switching histogram, we see that nearly twice as many trials resulted in a win than a loss. The not switching histogram shows that twice as many trials resulted in a loss than a win. From the trials we can get a higher probability of winning the car for switching, nearly 2/3 versus 1/3 for not switching. From this we can conclude that always switching produces more favorable odds or probabilities, and a contestant should always switch to increase their chance of winning the car.

## Simulation

Overall process for each trial:
1. Producers randomly place a car behind a door. We assume there are only 3 doors.
2. Contestant selects a door.
3. Host removes door without a car and not picked by contestant.
4. With two doors remaining, contestant is given the option to (a) switch or (b) stay.
5. Determine if contestant wins the car.

#### Door Removal

The following function executes door removal, step 3. Note these helper functions are presented a bit out of order because they need to be executed prior to running the main simulation.

```r
doorRemoval <- function(x, doorList = c(1,2,3)){ # x = row of door chosen and door with car.
  doorList <- as.data.frame(doorList)
  colnames(doorList) <- c("doors")
  # examine door chosen and door with car, get list of doors available for removal (no car, not chosen)
  doorList <- doorList[!(doorList$doors %in% as.list(x)),] 
  ifelse(length(doorList) == 1, 
            a <- doorList, # ex: if door chosen is 1 and winning door is 3, door removed = 2
            a <- sample(doorList, 1)) # ex: if door chosen is 3 and winning door is 3, door removed is either 1 or 2 at equal odds
  return(a)
}
```

#### Door Switch

The following function switches the contestant's door choice in step 4. 

```r
doorSwitching <- function(x, doorList = c(1,2,3)){ # x = row of door chosen, door win, and door removed
  doorList <- as.data.frame(doorList)
  colnames(doorList) <- c("doors")
  # examine door list and switch 
  doorList <- doorList[!(doorList$doors %in% as.list(x)),] 
  a <- doorList
  return(a)
}
```

#### Main Simulation

The following function executes steps 1, 2, and 5. Additionally, it outputs statistics and histograms in a consumable manner to evaluate which strategy to chose. **Note that both the no switch and switch strategy is combined into the same trial, but the calculation and evaluation is independent of each strategy. See discussion section below, if confused.**

```r
require(ggplot2)

sim.Monty <- function(n = 100, build.hist = FALSE){
  
  doors = c(1,2,3)
  # car randomly placed in one of three doors based on equal weights and probs.
  doorCar <- sample(doors, size = n, replace = TRUE, prob = c(1/3, 1/3, 1/3))
  # contestant picks one of three doors based on equal weights and probs.
  doorChoosen <- sample(doors, size = n, replace = TRUE, prob = c(1/3, 1/3, 1/3))
  simDoors <- data.frame(doorCar = doorCar, doorChoosen = doorChoosen)
  
  # host removes the door that doesn't have a car nor the one choosen by contestant.
  doorRemoved <- apply(as.matrix(simDoors), 1, doorRemoval, doorList = doors)
  simDoors <- cbind(simDoors, doorRemoved = doorRemoved)
  
  # scenario #1 - contestant always switches doors
  doorSwitched <- apply(simDoors[,2:3], 1, doorSwitching)
  simDoors <- cbind(simDoors, doorSwitched = doorSwitched)
  
  # scenario # 2 - contestant always stays on original choice
  doorNotSwitched <- apply(simDoors, 1, function(x) x[2])
  simDoors <- cbind(simDoors, doorNotSwitched = doorNotSwitched)
  
  # assess wins and losses for both scenarios 
  switchWin <- apply(simDoors, 1, function(x) ifelse(x[1]==x[4], 1, 0))
  noSwitchWin <- apply(simDoors, 1, function(x) ifelse(x[1]==x[5], 1,0))
  
  simDoors <- cbind(simDoors, switchWin = switchWin, noSwitchWin = noSwitchWin)  
  
  if (build.hist == TRUE){
    print(head(simDoors))  
    print(ggplot(simDoors, aes(x=switchWin)) + geom_bar() + ggtitle("Switching Frequencies"))
    print(ggplot(simDoors, aes(x=noSwitchWin)) + geom_bar() + ggtitle("No Switching Frequencies"))
    #barplot(simDoors$switchWin)
    #barplot(simDoors$noSwitchWin)
  }
  
  # build scenarios tables with frequencies, mean or probability of winning, 
  # variance, wins, losses, and iterations
  a <- c(mean(simDoors$switchWin), 
         var(simDoors$switchWin), 
         length(simDoors$switchWin[simDoors$switchWin == 0]), 
         length(simDoors$switchWin[simDoors$switchWin == 1]), n)
  
  b <- c(mean(simDoors$noSwitchWin), 
         var(simDoors$noSwitchWin), 
         length(simDoors$noSwitchWin[simDoors$noSwitchWin == 0]), 
         length(simDoors$noSwitchWin[simDoors$noSwitchWin == 1]), n)
  
  c <- as.data.frame(rbind(a, b))
  row.names(c) <- c("Switching", "No Switching")
  colnames(c) <- c("mean_winProb", "variance", "Losses", "Wins", "Iterations")
  
  return(c)
  
}
```

#### Run Simulation 

Running the simulation with 100,000 trials, we get several outputs. 

1. An overall structure of the trials (represented as rows) and steps mentioned previously. 

    - doorCar = door that the producers put the car behind (randomly chosen)
    - doorChosen = the door the contestant chooses (randomly chosen without knowledge of a)
    - doorRemoved = the door the host removes
    - doorSwithched = if taking the always switch mentality, what door contestant changes to
    - doorNOtSwitched = if taking the always stay mentality, the original door remains
    - SwitchWin = compare doorSwitch to doorCar, if the same then contestant wins (represented as 1)
    - NoSwitchWin = compare doorNotSwitched to doorCar, if the same the contestant wins
      
      
2. Histograms that show the frequencies of losses(0) to wins(1) in each strategy.
3. Table showing the mean or the probability of winning, variance, trials lost, trials won, and number of trials.


```r
table.sim <- sim.Monty(n = 100000, build.hist = TRUE)
```

```r
table.sim
```

## Discussion

The above histograms and table shows that the odds of winning a car if the contestant switches doors is 2:1 or a probability of 0.66976. The odds of winning a car if the contestant doesn't switch is 1:2 or a probability of 0.33024. The histograms that count wins and losses in the total trials show that for switching in the simulation resulted in nearly twice as many winining trials as losses. For not switching nearly twice as many trials resulted in a loss.  From this we can conclude that always switching is the best strategy because it increses your odds of winning.

The variance for both strategies is the same or similar, which tells us that the data is distributed similarly in both strategies (but around different means). This is expected since the winning results for switching are similar (the same in table above) to the losing results for not switching.  


### Note on combining switching and not switching into the same simulation

The win results in the table above are inverted for the two strategies (i.e. NoSwitchWins = 1e5 - SwitchWins), however, this doesn't affect the overall conclusion. If combining both strategies into the same trials is confusing  and you question the independence of the calculation, we can re-run the simulation twice and examine switching and not switching in separate tables. From the both tables below, we see that the win probabilites and counts are quite similar to those in the combined table above.

```r
table.simSwitch <- sim.Monty(n = 100000, build.hist = FALSE)
table.simSwitch[1,]
```
